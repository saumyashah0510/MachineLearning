## Filtering Data

```python
import pandas as pd
import numpy as np
```

```python
bios = pd.read_csv('bios.csv')
```

```python
bios.head()
```

```python
bios.loc[(bios["height_cm"] > 215) & (bios["born_country"] == 'USA'), ["name", "height_cm"]]
```

```python
bios[bios['name'].str.contains('keith|patrick',case=False)]
```

```python
bios[bios['born_country'].isin(['USA','GBR']) & (bios['name'].str.lower().str.startswith("keith"))]
```

```python
bios.query('born_country == "USA" and born_city == "Seattle"')
```

## Adding/Removing Columns

```python
coffee = pd.read_csv('coffee.csv')
coffee_og = coffee.copy()
```

```python
coffee["price"] = 4.99
coffee.head()
```

```python
coffee["new_price"] = np.where(coffee["Coffee Type"] == "Espresso", 5.99, 3.99)
coffee
```

```python
coffee.drop(columns=['price'],inplace=True)
coffee.head()
```

```python
coffee_og
```

```python
coffee['Revenue'] = coffee['Units Sold'] * coffee['new_price']
coffee.rename(columns={'new_price':'Price'}, inplace=True)
coffee.head()
```
