```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras.layers import Dense
from tensorflow.keras import Sequential
from tensorflow.keras.losses import MeanAbsoluteError
from tensorflow.keras.optimizers import SGD,Adam
from tensorflow.keras.callbacks import EarlyStopping
```

```python
X = tf.range(-100,100,4)
y = X + 10
plt.scatter(X,y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X.numpy(), 
                                                    y.numpy(), 
                                                    test_size=0.2,
                                                    random_state=42,
                                                    shuffle=False)
```

## Model 1 :

One layer 100 epochs

```python
from tabnanny import verbose


tf.random.set_seed(42)

model_1 = Sequential(
    [
        Dense(1,input_shape=(1,))
    ]
)

model_1.compile(loss = MeanAbsoluteError,
                optimizer = SGD(),
                metrics = ["MeanAbsoluteError"])

model_1.fit(X_train,y_train,epochs=100,verbose=0)

y_pred_1 = model_1.predict(X_test)

mae_1 = MeanAbsoluteError()(y_test, y_pred_1)
mse_1 = tf.keras.losses.MeanSquaredError()(y_test, y_pred_1)
```

## Model 2 :

2 layers, 100 epochs

```python
tf.random.set_seed(42)

model_2 = Sequential(
    [
        Dense(10,input_shape=(1,)),
        Dense(1)
    ]
)

model_2.compile(loss = MeanAbsoluteError,
                optimizer = SGD(),
                metrics = ["MeanAbsoluteError"])

model_2.fit(X_train,y_train,epochs=100)

y_pred_2 = model_2.predict(X_test)
plot_predictions(predictions=y_pred_2)

mae_2 = MeanAbsoluteError()(y_test, y_pred_2)
mse_2 = tf.keras.losses.MeanSquaredError()(y_test, y_pred_2)
```

## Model 3:
2 layers and 500 epochs

```python
tf.random.set_seed(42)

model_3 = Sequential(
    [
        Dense(10,input_shape=(1,)),
        Dense(1)
    ]
)

model_3.compile(loss = MeanAbsoluteError,
                optimizer = SGD(),
                metrics = ["MeanAbsoluteError"])

model_3.fit(X_train,y_train,epochs=500)

y_pred_3 = model_3.predict(X_test)
plot_predictions(predictions=y_pred_3)

mae_3 = MeanAbsoluteError()(y_test, y_pred_3)
mse_3 = tf.keras.losses.MeanSquaredError()(y_test, y_pred_3)
```

```python
tf.random.set_seed(42)

model_4 = Sequential(
    [
        Dense(1,input_shape=(1,))
    ]
)

early_stopping = EarlyStopping(monitor='loss', patience=50)

model_4.compile(loss = MeanAbsoluteError,
                optimizer = Adam(learning_rate=0.1),
                metrics = ["MeanAbsoluteError"])

model_4.fit(X_train,y_train,epochs=500,callbacks=[early_stopping])

y_pred_4 = model_4.predict(X_test)
plot_predictions(predictions=y_pred_4)

mae_4 = MeanAbsoluteError()(y_test, y_pred_4)
mse_4 = tf.keras.losses.MeanSquaredError()(y_test, y_pred_4)
```
