```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras.layers import Dense
from tensorflow.keras import Sequential
from tensorflow.keras.losses import MeanAbsoluteError
from tensorflow.keras.optimizers import SGD,Adam
from tensorflow.keras.callbacks import EarlyStopping
```

```python
X = np.array([[-7.0], [-4.0], [-1.0], [2.0], [5.0], [8.0], [11.0], [14.0]])
y = np.array([[3.0], [6.0], [9.0], [12.0], [15.0], [18.0], [21.0], [24.0]])

plt.scatter(X,y)
```

```python
input_shape = X.shape
output_shape = y.shape
input_shape,output_shape
```

```python
tf.random.set_seed(42)

model = Sequential(
    [
        Dense(1,input_shape=(1,)),
    ]
)

model.compile(loss = MeanAbsoluteError(),
              optimizer = Adam(learning_rate=0.1),
              metrics = ["MeanAbsoluteError"])

model.fit(X,y,epochs=500,callbacks=[EarlyStopping()])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X.numpy(), 
                                                    y.numpy(), 
                                                    test_size=0.2,
                                                    random_state=42)
    
model.fit(X_train,y_train,epochs=100,callbacks=[EarlyStopping()])
```

```python
y_pred = model.predict(X_test)
```

```python
def plot_predictions(train_data, train_labels, test_data, test_labels, predictions):
  """
  Plots training data, test data, and compares predictions.
  """
  plt.figure(figsize=(10, 5))

  
  plt.subplot(1, 2, 1) 
  plt.title("Model Training")
  plt.scatter(train_data, train_labels, c="blue", label="Training data")
  plt.plot(test_data, predictions, c="red", linewidth=3, label="Model's Fit") # The line shows the model
  plt.xlabel("X")
  plt.ylabel("y")
  plt.legend()


  plt.subplot(1, 2, 2)
  plt.title("Model Predictions vs. Actual")
  plt.scatter(test_data, test_labels, c="green", label="Actual Test Values")
  plt.scatter(test_data, predictions, c="orange", marker='x', label="Predicted Values")
  plt.xlabel("X")
  plt.ylabel("y")
  plt.legend()

  plt.tight_layout() 
  plt.show()

plot_predictions(X_train, y_train, X_test, y_test, y_pred)  
```
