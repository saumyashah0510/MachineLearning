## Splitting Data 

```python
from sklearn.datasets import load_iris # Flowers
from sklearn.model_selection import train_test_split
```

```python
data = load_iris()
X,y = data.data,data.target
X
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2) # 20% testing and 80% training
print(len(X_train),len(X_test))
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
counts = np.bincount(y_train)
positions = np.arange(3)

plt.bar(positions,counts)
plt.xticks(positions,data.target_names)
plt.show()
```

```python
counts = np.bincount(y_test)
positions = np.arange(3)

plt.bar(positions,counts)
plt.xticks(positions,data.target_names)
plt.show()
```

### Stratified shuffle split :

Equally split training and testing i.e. keeps proportions same

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1,test_size=0.2)

for train_idx , test_idx in split.split(X,y):
    X_train,X_test = X[train_idx],X[test_idx]
    y_train,y_test = y[train_idx],y[test_idx]
```

```python
counts = np.bincount(y_train)
positions = np.arange(3)

plt.bar(positions,counts)
plt.xticks(positions,data.target_names)
plt.show()
```

```python
counts = np.bincount(y_test)
positions = np.arange(3)

plt.bar(positions,counts)
plt.xticks(positions,data.target_names)
plt.show()
```
