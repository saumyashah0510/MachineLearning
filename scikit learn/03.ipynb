## PreProcessing

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X,y = load_iris(return_X_y=True)
```

```python
X_train , X_test , y_train,y_test = train_test_split(X,y,test_size=0.2)
```

### Standard Scaling

```python
scalar = StandardScaler()

X_train_scaled = scalar.fit_transform(X_train)
X_test_scaled = scalar.transform(X_test)
```

```python
X_train_scaled
```

### Manual Standard Scaling

```python
import numpy as np

(X_train - np.mean(X_train,axis=0))/np.std(X_train,axis=0)
```

---
### Min-Max Scaling

```python
from sklearn.preprocessing import MinMaxScaler

scaler_min_max = MinMaxScaler()

X_train_scaled_minmax = scaler_min_max.fit_transform(X_train)
X_test_scaled_minmax = scaler_min_max.transform(X_test)

X_train_scaled_minmax
```

### Manual MinMax Scaling

```python
X_min = np.min(X_train,axis = 0)
X_max = np.max(X_train,axis = 0)

(X_train-X_min)/(X_max - X_min)
```

## Feature Decoding

---
### Ordinal Decoding : 
When there is progression and we can order it as 1,2,3

```python
from sklearn.datasets import fetch_openml
from sklearn.preprocessing import OrdinalEncoder
```

```python
data = fetch_openml('car',as_frame=True).frame
```

```python
data
```

```python
columns_to_encode = ['lug_boot','safety']

encoder = OrdinalEncoder(
    categories=[
        ['small','med','big'],
        ['low','med','high']
    ]
)

data[columns_to_encode] = encoder.fit_transform(data[columns_to_encode])
```

```python
data
```

```python
encoder.inverse_transform(data[columns_to_encode]) # To reverse encoding
```
