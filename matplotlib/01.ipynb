# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4]
y = [10,20,15,25]
plt.plot(x,y)
plt.show()
```

```python
x = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']
y = [10,15,7,20,12,30,25]
plt.plot(x,y)
plt.title('Bakery Sales This Week')
plt.xlabel('Day of the week')
plt.ylabel('Sales per day')
plt.show()
```

```python

months = [1,2,3,4]
sales = [1000,1500,1200,1800]

plt.plot(months,sales,color='red',linestyle = '--', linewidth=2,marker='o',label='2025 Sales Data')
plt.title('Monthly Sales Data Report')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend(loc='upper left', fontsize='small')
plt.xlim([1, 4])
plt.ylim([1000, 2000])
plt.xticks([1, 2, 3, 4], ['M1', 'M2', 'M3', 'M4'])
plt.grid()
plt.show() 
```

```python
product = ['A', 'B', 'C', 'D']
sales = [1000,1500,800,1200]

plt.bar(product,sales,color = 'orange', label = "Sales 2025")
plt.xlabel('Product')
plt.ylabel('Sales')
plt.title('Product Sales Data Report')
plt.legend()
plt.show()
```

```python
regions = ['North', 'South', 'East', 'West']
revenue = [2500, 3000, 1500, 2000]

plt.pie(revenue, labels=regions, autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'lightcoral', 'gold'])
plt.title('Revenue Distribution by Region')
plt.show()
```

```python
scores = [57,89,67,45,78,90,88,76,95,84,91,73,60]
plt.hist(scores,bins=5,color='pink',edgecolor='black')
plt.xlabel('Score ranges')
plt.ylabel('Number of Students')
plt.title('Distribution of Student Scores')
plt.show()
```

```python
hours_studied = [1, 2, 3, 4, 5, 6, 7, 8]
scores = [50, 55, 60, 65, 70, 75, 80, 85]
plt.scatter(hours_studied, scores, color='red', marker='o', label='Student Scores')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Hours Studied vs Exam Score')
plt.legend()    
plt.grid(True)
plt.show()
```

```python

plt.scatter([1,2,3],[50,55,60], color='red', marker='o', label='Class A')
plt.scatter([1,2,3],[45,50,52], color='blue', marker='o', label='Class B')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Comparision of two classes')
plt.legend()    
plt.grid(True)
plt.show()
```

```python
x = [1,2,3,4]
y = [10,20,15,25]

plt.subplot(1,2,1)  # 1 row, 2 columns, first subplot
plt.plot(x,y)
plt.title('Line Plot')

plt.subplot(1,2,2)  # 1 row, 2 columns, second subplot
plt.bar(x,y)
plt.title('Bar Chart')

plt.show()
```

```python
x = [1,2,3,4]
y = [10,20,15,25]

fig,ax = plt.subplots(1,2,figsize=(10,5))  # 1 row, 2 columns

ax[0].plot(x,y)
ax[0].set_title('Line Plot')

ax[1].bar(x,y)
ax[1].set_title('Bar Chart')

fig.suptitle('Line vs Bar Chart')

plt.tight_layout()
plt.show()
```
